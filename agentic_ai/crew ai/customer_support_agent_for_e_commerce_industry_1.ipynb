```python
from langchain_cohere import ChatCohere

# Use Cohere Command-R (chat model) or Command-Light
llm = ChatCohere(model="command-r-plus", temperature=0.3)
```

```python
from crewai import Agent, Task, Crew

# Agents
support_agent = Agent(
    role="Customer Support Agent",
    goal="Help customers with orders, refunds, and product info",
    backstory="An AI assistant for an e-commerce company",
    llm=llm
)

faq_agent = Agent(
    role="Knowledge Agent",
    goal="Provide accurate answers from FAQ and policy knowledge base",
    backstory="An AI agent that knows company policies and FAQs",
    llm=llm
)

escalation_agent = Agent(
    role="Escalation Agent",
    goal="Detect unresolved issues and suggest escalation to human agent",
    backstory="Routes tough cases to human support",
    llm=llm
)

# Tasks
faq_task = Task(
    description="Answer customer query using FAQ or policies",
    expected_output="Clear, policy-based answer",
    agent=faq_agent
)

support_task = Task(
    description="Respond empathetically to customer queries, using FAQ or order info",
    expected_output="Polite support response",
    agent=support_agent
)

escalation_task = Task(
    description="Decide if issue needs escalation to human support",
    expected_output="Escalation decision",
    agent=escalation_agent
)

# Crew
crew = Crew(
    agents=[support_agent, faq_agent, escalation_agent],
    tasks=[faq_task, support_task, escalation_task],
    verbose=True
)

# Run
query = "I want to return a product I bought last week."
result = crew.kickoff(inputs={"query": query})

print("âœ… Final Answer:", result)
```

```python
from crewai import Agent, Task, Crew
from langchain_cohere import ChatCohere

# Initialize Cohere LLM
llm = ChatCohere(model="command-r-plus", temperature=0.3)

# Agents
support_agent = Agent(
    role="Customer Support Agent",
    goal="Assist customers with orders, refunds, returns, and product info",
    backstory="An AI support agent for an e-commerce company",
    llm=llm
)

faq_agent = Agent(
    role="Knowledge Agent",
    goal="Answer FAQs and provide policy/product info",
    backstory="Specialist in FAQs and company policies",
    llm=llm
)

escalation_agent = Agent(
    role="Escalation Agent",
    goal="Detect unresolved or sensitive issues and escalate",
    backstory="Routes difficult cases to human support",
    llm=llm
)

# Tasks
faq_task = Task(
    description="Answer customer query using FAQ or policies",
    expected_output="A clear FAQ-based response",
    agent=faq_agent
)

support_task = Task(
    description="Respond empathetically to customer queries",
    expected_output="Polite customer support response",
    agent=support_agent
)

escalation_task = Task(
    description="Decide if the issue needs escalation",
    expected_output="Escalation decision",
    agent=escalation_agent
)

# Crew Manager
crew = Crew(
    agents=[support_agent, faq_agent, escalation_agent],
    tasks=[faq_task, support_task, escalation_task],
    verbose=False
)
```

```python
import gradio as gr

def customer_support_chat(user_input, history):
    result = crew.kickoff(inputs={"query": user_input})
    return str(result)

with gr.Blocks() as demo:
    gr.Markdown("## ğŸ›’ E-Commerce Customer Support Chatbot (Cohere + CrewAI)")
    chatbot = gr.ChatInterface(fn=customer_support_chat)

demo.launch(share=True)  # share=True gives you a public link
```
