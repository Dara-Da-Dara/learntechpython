```python
from llama_index.embeddings.cohere import CohereEmbedding

# -------------------------------
# Step 1: Initialize Cohere embedding model
# -------------------------------
embed_model = CohereEmbedding(
    api_key="api key ",
    model_name="embed-english-v3.0",
    input_type="search_document",  # Use "search_query" for queries, "search_document" for documents
    max_tokens=8000,
    embedding_types=["float"],
)

# -------------------------------
# Step 2: 30-line text about Pankaj Pandey
# -------------------------------
text_to_embed = """
Pankaj Pandey is an accomplished data analyst with extensive experience in analytics and business intelligence.
He has over five years of experience working in IT and data-driven companies.
Pankaj has expertise in Python, SQL, and R for data manipulation and analysis.
He is highly skilled in building interactive dashboards using Power BI and Tableau.
Pankaj has worked on multiple projects involving predictive modeling and forecasting.
He has a strong background in statistical analysis and machine learning techniques.
Pankaj is proficient in cloud platforms such as AWS and Microsoft Azure.
He is experienced in ETL processes, data cleaning, and pipeline automation.
Pankaj has a proven track record of delivering actionable insights to business stakeholders.
He is highly adept at working in cross-functional teams and collaborating effectively.
Pankaj is passionate about leveraging data to improve business outcomes.
He has experience mentoring junior analysts and leading small teams.
Pankaj has successfully implemented automated reporting systems in his previous roles.
He is recognized for improving operational efficiency through data-driven solutions.
Pankaj is detail-oriented and committed to delivering high-quality work consistently.
He has worked on projects involving customer analytics and market trend analysis.
Pankaj is familiar with Agile and Scrum methodologies for project management.
He has experience designing dashboards that are both visually appealing and informative.
Pankaj continuously updates his skills to stay current with the latest data technologies.
He has successfully reduced manual reporting efforts through automation.
Pankaj is highly organized and capable of managing multiple projects simultaneously.
He has experience working with large datasets and optimizing queries for performance.
Pankaj is a strong communicator and able to explain complex data insights clearly.
He has contributed to revenue forecasting and business growth initiatives.
Pankaj has received recognition for his contributions to team projects and business solutions.
He is committed to continuous learning and professional development in analytics.
Pankaj has experience integrating multiple data sources for comprehensive analysis.
He is skilled at identifying patterns, trends, and actionable insights from data.
Pankaj is motivated by challenges and enjoys solving complex analytical problems.
He aspires to leverage his data expertise to make a significant impact in business decision-making.
"""

# -------------------------------
# Step 3: Define questions
# -------------------------------
questions = [
    "Who is Pankaj Pandey and what is his profession?",
    "How many years of experience does Pankaj Pandey have in IT and analytics?",
    "What programming languages is Pankaj Pandey skilled in?",
    "Which data visualization tools does Pankaj Pandey use?",
    "What types of projects has Pankaj Pandey worked on?",
    "What cloud platforms is Pankaj Pandey proficient with?",
    "How has Pankaj Pandey contributed to business intelligence or decision-making?",
    "Which methodologies is Pankaj Pandey familiar with in project management?",
    "Can you list some key achievements of Pankaj Pandey?",
    "What qualities make Pankaj Pandey an effective team member or leader?"
]

# -------------------------------
# Step 4: Embed the document once
# -------------------------------
document_embedding = embed_model.get_text_embedding(text_to_embed)

# -------------------------------
# Step 5: Function to answer a query
# -------------------------------
import numpy as np

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

def answer_question(query):
    # Embed the query
    query_embedding = embed_model.get_text_embedding(query)

    # Compute similarity with the document
    similarity_score = cosine_similarity(query_embedding, document_embedding)

    # Create a prompt for the LLM
    prompt = f"""
You are an assistant with knowledge about Pankaj Pandey.
Use the information below to answer the question accurately.

Information:
{text_to_embed}

Question:
{query}

Answer:
"""
    return prompt, similarity_score

# -------------------------------
# Step 6: Generate prompts and similarity for each question
# -------------------------------
for q in questions:
    prompt, score = answer_question(q)
    print(f"Question: {q}")
    print(f"Similarity Score: {score:.4f}")
    print("Generated Prompt for LLM:")
    print(prompt)
    print("="*80)
```
