```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

# Load model (code generation model)
model_name = "Salesforce/codegen-350M-mono"  # You can use other code models like codegen-2B-multi
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

def generate_code(prompt, max_length=200):
    inputs = tokenizer(prompt, return_tensors="pt")
    outputs = model.generate(
        **inputs,
        max_length=max_length,
        do_sample=True,
        temperature=0.7,
        top_p=0.95,
        top_k=50
    )
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Example usage
prompt_python = "Write a Python function to calculate factorial of a number"
prompt_java = "Write a Java class to reverse a string"
prompt_react = "Write a React component that shows a button and on click shows alert"
prompt_html = "Write an HTML page with a form that has name and email input fields"

print("Python Code:\n", generate_code(prompt_python))
print("\nJava Code:\n", generate_code(prompt_java))
print("\nReact Code:\n", generate_code(prompt_react))
print("\nHTML Code:\n", generate_code(prompt_html))
```
