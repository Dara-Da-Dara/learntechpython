```python
from groq import Groq
import json

client = Groq(api_key="api")

tools = [
    {
        "type": "function",
        "function": {
            "name": "get_stock_price",
            "description": "Get the latest price of a stock by symbol",
            "parameters": {
                "type": "object",
                "properties": {
                    "symbol": {
                        "type": "string",
                        "description": "Stock symbol, e.g. AAPL, MSFT, TSLA"
                    }
                },
                "required": ["symbol"],
            },
        },
    }
]

def get_stock_price(symbol: str) -> float:
    fake_prices = {"AAPL": 223.45, "MSFT": 415.67, "TSLA": 245.32}
    return fake_prices.get(symbol, 0.0)

available_functions = {"get_stock_price": get_stock_price}

messages = [
    {"role": "user", "content": "What is the price of AAPL right now?"}
]

# Use Groq’s recommended, non-deprecated model
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",   # or "llama-3.3-70b-versatile"
    messages=messages,
    tools=tools,
    tool_choice="auto",
)

msg = response.choices[0].message
tool_calls = getattr(msg, "tool_calls", None)

if tool_calls:
    messages.append(msg)
    for tc in tool_calls:
        name = tc.function.name
        args = json.loads(tc.function.arguments)
        result = available_functions[name](**args)
        messages.append(
            {
                "role": "tool",
                "tool_call_id": tc.id,
                "name": name,
                "content": str(result),
            }
        )

    final = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages,
    )
    print(final.choices[0].message.content)
else:
    print(msg.content)
```

```python
This appears to be the current price of AAPL (Apple Inc.) in my training data. However, this is not real-time data, and it might be outdated.

For real-time prices, I recommend checking a reliable financial website, such as Yahoo Finance, Google Finance, or Bloomberg, or using a financial API.
```
